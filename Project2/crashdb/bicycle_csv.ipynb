### This notebook transforms the biker data csv. 

* Un-needed columns removed


```python
import pandas as pd
import pymongo
import json
```

```python
# Imported file into Pandas

file = "bicycle_data.csv"

df = pd.read_csv(file, low_memory=False)
df.head()
```

```python
# Removing un-needed columns

dropped_df = df.drop(["Geo Point", "Geo Shape", "Ambulance", "Biker Age Group", "Biker Age", "Biker Injury", "Bike Direction", "Bike Position", "Biker Race", "Biker Sex", "City", "County", "Day", "Cause of Incident", "Location of Crash", "Month", "Time (Military Time)", "Crash Type", "Year", "Development", "Driver Age", "Driver Injury", "Driver Race", "Driver Sex", "Driver Vehicle Type", "Excessive Speeding", "Hit and Run", "Lighting Conditions", "Locality", "Number of Lanes", "Number of Units", "Road Characteristics", "Road Class", "Road Configuration", "Road Defects", "Road Feature", "Road Surface", "Rural/Urban", "Speed Limit", "Traffic Control", "Weather", "Workzone"], axis=1)
dropped_df.head()
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.crash_db
collection = db.biker_information
```

```python
records = json.loads(dropped_df.T.to_json()).values()
```

```python
db.biker_information.insert_many(records)
```
